```python
import numpy as np
import pandas as pd

def count_s(df,st): #func to count strings in a matrix
    return sum([list(df[i]).count(st) for i in df])

def slicing(df,i,j,c,t): 
    return df.loc[i:j,c:t]
def Multiples(value): #to get possible vectors
    return [[i,value/i] for i in range(1,value+1) if value %i==0]
```

```python
slices=[]
startpos=[0,0]
while(1<2):
    for dim in poss:
        t=1
        if (startpos[0]+dim[0]-1<=int(header[0]) and startpos[1],startpos[1]+dim[1]-1<=int(header[1])):
            my_slice=slicing(data,startpos[0],startpos[0]+dim[0]-1,startpos[1],startpos[1]+dim[1]-1)
            if count_s(my_slice,"T") >= int(header[2]) and count_s(my_slice,"M") >= int(header[2]):
                t=0
                break
    if t==0:
        data.loc[startpos[0]:startpos[0]+dim[0]-1,startpos[1]:startpos[1]+dim[1]-1]="NaN" #converting pieces to NaN
        slices.append([startpos[0],startpos[0]+dim[0]-1,startpos[1],startpos[1]+dim[1]-1])
    if startpos[1]+dim[1]-1<int(header[1]): # the bug starts here its about picking the next start point!
        startpos[1]=startpos[1]+dim[1]
    elif startpos[0]+dim[0]-1<int(header[0]):
        startpos[1]=0
        startpos[0]=startpos[0]+dim[0]
    else:
        break
```
